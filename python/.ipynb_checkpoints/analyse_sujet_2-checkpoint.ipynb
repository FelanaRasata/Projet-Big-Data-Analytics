```python
from pyhive import hive
import pandas as pd
from IPython.display import display
```

1. Paramètres de connexion Hive 

```python
hive_host = 'localhost'
hive_port = 10000
hive_username = ' ' 
hive_password = ' '
```

```python
#conn = None
try:
    # Établir une connexion avec authentification LDAP
    conn = hive.Connection(
        host=hive_host,
        port=hive_port,
        username=hive_username,
        password=hive_password,
        auth='LDAP'  
    )
    print("Connecté à Hive avec succès")
except Exception as e:
    print(f"Erreur lors de la connexion à Hive: {e}")
```

2. Chargement de données

```python
# Créer un curseur
cursor=conn.cursor()
```

```python
def findAll(table) :
    # Exécuter une requête pour récupérer les données de la table "catalogue"
    query = "SELECT * FROM " + table
    cursor.execute(query)

    # Charger les résultats dans un DataFrame Pandas
    data = cursor.fetchall()

    # Récupérer les noms des colonnes
    columns = [desc[0] for desc in cursor.description] 
    df = pd.DataFrame(data, columns=columns)
    print("Select all ",table)

    display(df.head())
    
    return df
```

```python
dfCatalogue = findAll("catalogue_co2_ext")
```

    - Immatriculation

```python
dfImmatriculation = findAll("immatriculation_ext")
```

    - Marketing 

```python
dfMarketing = findAll("marketing_ext")
```

    - Client

```python
dfClients = findAll("clients_ext")
```

## Statistiques descriptives / vérification des valeurs

```python
def descriptionDataframe(table, df) :
    # Affichage des statistiques descriptives
    print("Affichage des statistiques descriptives de ", table)
    display(df.describe())
    
    # Affichage des types de données et des valeurs manquantes
    print("Affichage des types de données et des valeurs manquantes  de ", table)
    display(df.info())

    # Vérification des valeurs manquantes
    print("Vérification des valeurs manquantes  de ", table)
    display(df.isnull().sum())
    
    # Vérification des doublons
    print("Vérification des doublons  de ", table)
    duplicates = df.duplicated()
    print(f"Nombre de doublons : {duplicates.sum()}")
```

## Histogramme / Nuage de points / Matrice de corelaton

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Exclure les colonnes non numériques spécifiées
df = dfCatalogue.copy()
df['catalogue_co2_ext.longueur'] = df['catalogue_co2_ext.longueur'].map({'très longue': 3, 'longue': 2, 'moyenne': 1, 'courte': 0})

# Calcul de la matrice de corrélation
corr_matrix = df[[
    'catalogue_co2_ext.puissance', 
    'catalogue_co2_ext.nbplaces', 
    'catalogue_co2_ext.nbportes', 
    'catalogue_co2_ext.prix', 
    'catalogue_co2_ext.longueur',
    'catalogue_co2_ext.bonusmalus',
    'catalogue_co2_ext.rejetco2',
    'catalogue_co2_ext.coutenergie'
]].corr()
display(corr_matrix)

# Visualisation de la matrice de corrélation
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap des corrélations')
plt.show()
```

```python
# Affichage des histogrammes pour les colonnes numériques
for col in [
    'catalogue_co2_ext.puissance', 
    'catalogue_co2_ext.nbplaces', 
    'catalogue_co2_ext.nbportes', 
    'catalogue_co2_ext.prix', 
    'catalogue_co2_ext.longueur',
    'catalogue_co2_ext.bonusmalus',
    'catalogue_co2_ext.rejetco2',
    'catalogue_co2_ext.coutenergie'
]:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogramme de {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.show()
```

### Conclusion:

1. Variables importantes:

   - La puissance, le prix et la longueur sont des variables importantes car elles montrent de fortes corrélations avec plusieurs autres variables. Elles peuvent être utilisées pour distinguer les catégories de véhicules.

   - Le nombre de places, le rejet CO2, le bonus et le cout d'énergie montrent peu de corrélations significatives avec les autres variables,mais ils ne doivent pas être exclu sans une analyse plus approfondie de son impact potentiel sur la catégorisation.

   - Le rejet CO2, le bonus et le cout d'énergie sont des variables très correlés alors on va considerer un seul de ces variables.

3. Variables moins pertinentes: nbportes semble avoir une distribution homogène et pourrait être moins informative pour le clustering.

4. Variables à considérer pour le Clustering: puissance, nbplaces, prix, longueur, coutenergie.

# 2. Categorisation des véhicules

```python
col_predictive = [
    'catalogue_co2_ext.puissance', 
    'catalogue_co2_ext.nbportes', 
    'catalogue_co2_ext.prix', 
    'catalogue_co2_ext.longueur',
    'catalogue_co2_ext.rejetco2'
]
```

```python
# Sélectionner les caractéristiques pertinentes
dfCatalogue_used = dfCatalogue[col_predictive]
dfCatalogue_used.head(5)
```

```python
# Attribuer des valeurs numériques spécifiques aux autres variables catégorielles
category_mapping = {
    'très longue': 3,
    'longue': 2,
    'moyenne': 1,
    'courte': 0
}

for col in col_predictive:
    if dfCatalogue_used[col].dtype == 'object':
        dfCatalogue_used[col] = dfCatalogue_used[col].map(category_mapping)

display(dfCatalogue_used.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Normalisation des données
scaler = StandardScaler()
scaled_dfCatalogue_used = scaler.fit_transform(dfCatalogue_used)
print("done")
```

```python
#remettre en dataframe le tableau numpy des données catalogué mise à l'echelle pour vérifier
scaled_dfCatalogue_used_toDf = pd.DataFrame(scaled_dfCatalogue_used, columns=dfCatalogue_used.columns)
print("done")

# Print the first few rows of the scaled DataFrame
display(scaled_dfCatalogue_used_toDf.head())
```

```python
from sklearn.cluster import KMeans

# Utiliser la méthode du coude pour trouver le nombre optimal de clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_dfCatalogue_used)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Méthode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('WCSS')
plt.show()
```

D'après la methode de coude, on voit ici que le nombre optimale de cluster est aux alentours de 4

```python
# Application de k-means avec le nombre de clusters choisi
kmeans = KMeans(n_clusters=4, random_state=42)
dfCatalogue['catalogue_co2_ext.cluster'] = kmeans.fit_predict(scaled_dfCatalogue_used)
print("done")
```

```python
# Calculer les caractéristiques moyennes de chaque cluster
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_labels = dfCatalogue['catalogue_co2_ext.cluster'].unique()
cluster_summary = pd.DataFrame(cluster_centers, columns=dfCatalogue_used.columns, index=cluster_labels)

display(cluster_summary)
```

```python
cluster_labels = {
    0: 'Citadine',
    1: 'Luxe',
    2: 'Compacte',
    3: 'Familiale'
}
dfCatalogue['catalogue_co2_ext.categorie'] = dfCatalogue['catalogue_co2_ext.cluster'].map(cluster_labels)

# Afficher le catalogue avec les catégories assignées
dfCatalogue.head()
```

# 3. Application des catégories de véhicules définies aux données des Immatriculations

```python
result_df = dfCatalogue.groupby('catalogue_co2_ext.marque', as_index=False).agg({'catalogue_co2_ext.rejetco2': 'mean'})
result_df.head()
```

```python
dfImmatriculation = pd.merge(
    dfImmatriculation, 
    result_df, how='inner', 
    left_on='immatriculation_ext.marque', 
    right_on='catalogue_co2_ext.marque'
)
```

```python
dfImmatriculation.head()
```

```python
col_predictive_immatriculation = [
    'immatriculation_ext.puissance', 
    'immatriculation_ext.longueur', 
    'immatriculation_ext.nbportes', 
    'immatriculation_ext.prix',
    'catalogue_co2_ext.rejetco2'
]
```

```python
# Sélectionner les caractéristiques pertinentes pour les immatriculations
dfImmatriculation_used = dfImmatriculation[col_predictive_immatriculation]

# Attribuer des valeurs numériques spécifiques aux autres variables catégorielles
category_mapping_2 = {
    'très longue': 3,
    'longue': 2,
    'moyenne': 1,
    'courte': 0
}
for colo in col_predictive_immatriculation:
    if dfImmatriculation_used[colo].dtype == 'object':
        dfImmatriculation_used[colo] = dfImmatriculation_used[colo].map(category_mapping_2)

dfImmatriculation_used.head()
```

```python
# Normalisation des données
scalerImma = StandardScaler()
scaled_dfImmatriculation_used = scalerImma.fit_transform(dfImmatriculation_used)
print("done scaling")
```

```python
# Utiliser le modèle de clustering pour prédire les clusters des données d'immatriculation
dfImmatriculation['immatriculation_ext.cluster'] = kmeans.predict(scaled_dfImmatriculation_used)

# Mapper les clusters prédits aux catégories de véhicules correspondantes
cluster_labels = {
    0: 'Citadine',
    1: 'Routière',
    2: 'Familiale',
    3: 'Sportive'
}
dfImmatriculation['immatriculation_ext.categorie'] = dfImmatriculation['immatriculation_ext.cluster'].map(cluster_labels)
print("done")

# Afficher le DataFrame d'immatriculation avec les catégories assignées
dfImmatriculation.head()
```

# 4. Fusion des données Clients et Immatriculations

```python
# Fusionner les deux DataFrames sur la colonne 'immatriculation'
dfFusionClientImmatricumation = pd.merge(dfClients, dfImmatriculation, how='inner', left_on='clients_ext.immatriculation', right_on='immatriculation_ext.immatriculation')
print("Fusion done")

# Afficher les premières lignes du DataFrame fusionné
dfFusionClientImmatricumation.head()
```
